```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'

def seniority(title):
    if 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'junior' in title.lower():
        return 'junior'
    else:
        return 'na'
```

```python
df['job_simplified'] = df['Job Title'].apply(title_simplifier)
```

```python
df.head()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
```

```python
df['job_state'] = df['job_state'].apply(lambda x : 'CA' if 'los angeles' in x.strip().lower() else x.strip())
```

```python
df.job_state.value_counts()
```

```python
df.head()
```

```python
# define job description length
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
```

```python
df['comp_counts'] = df['Competitors'].apply(lambda x: 0 if x == '-1' else len(x.split(',')))
```

```python
# df hourly wage to annual 
df['new_min_salary'] = df.apply(lambda x: x.min_salary * 2 if x.hourely == 1 else x.min_salary, axis = 1)
```

```python
df[['min_salary', 'hourely', 'new_min_salary']][df.hourely == 1]
```

```python
df.columns
```

```python
df['min_salary'] = df.apply(lambda x: x.min_salary * 2 if x.hourely == 1 else x.min_salary, axis = 1)
df['max_salary'] = df.apply(lambda x: x.max_salary * 2 if x.hourely == 1 else x.min_salary, axis = 1)
df['avg_salary'] = df.apply(lambda x: x.avg_salary * 2 if x.hourely == 1 else x.min_salary, axis = 1)
```

```python
df = df.drop('new_min_salary', axis = 1)
```
